# A Python Interface to the Master Hearing Aid (MHA)

## The actual interface class

The class is implemented as a `telnetlib.Telnet` subclass that provides additional higher-level methods for sending and receiving data from the MHA.

```python
import telnetlib
from collections import MutableSequence
import numpy as np
from ast import literal_eval
from encodings.utf_8 import encode as encode_utf8

class MHAConnection(telnetlib.Telnet):
    
    def send_command(self, cmd):
        
        self.write(cmd)
        err_code, _match, resp = self.expect([b'\(MHA:success\)', b'\(MHA:failure\)'])
        if err_code == 0:
            return resp.rpartition(b'(MHA:success)')[0].strip()
        else:
            raise ValueError('Error code {}:\n\t{}'.format(err_code, resp))
    
    def get_val(self, path):
        
        return self.send_command(path.strip() + b'?val\n')

    def get_type(self, path):
        
        return self.send_command(path.strip() + b'?type\n')
    
    def is_writable(self, path):
        
        ret = self.send_command(path.strip() + b'?perm\n')
        return ret == b'writable'
    
    def get_val_converted(self, path):
        
        data_type = self.get_type(path)
        data = self.get_val(path)
        
        if b'vector' in data_type or b'matrix' in data_type:
            data = data.replace(b' ', b', ').replace(b';', b',')

        # literal_eval() is a safe version of eval() that only accepts a small list of literal structures
        return literal_eval(data.decode())
    
    def set_val(self, path, value):
        
        cmd = path.strip() + b'='+ encode_utf8(value.strip())[0] + b'\n'
        return self.send_command(cmd)
    
    def set_val_converted(self, path, value):
        
        if isinstance(value, str):
            pass  # nothing to do
        elif isinstance(value, (np.ndarray, MutableSequence)):
            if isinstance(value, np.ndarray):
                if value.ndim > 2:
                    raise ValueError('"value" cannot have more than 2 dimensions')
                value = value.tolist()
            value = str(value).replace('],', '];').replace(',', '')
        else:
            value = str(value)  # hope for the best...
               
        return self.set_val(path, value)
```

Below is a button that toggles the connection to the MHA. Successive pressing of the button switches the connection off and on.

```python
from IPython.html import widgets # Widget definitions
from IPython.display import display # Used to display widgets in the notebook

conn = MHAConnection('127.0.0.1', 33337)

def connect_to_mha(Connect):
    if Connect and conn.sock == 0:
        conn.open('127.0.0.1', 33337)
        print('Connected to MHA')
    elif not Connect and conn.sock != 0:
        conn.close()
        print('Closed connection to MHA')
widgets.interact(
    connect_to_mha,
    Connect=widgets.ToggleButtonWidget(description="Connect", value=True)
);

def quit_mha(val):
    if conn.sock != 0:
        conn.send_command(b'cmd=quit\n')

quit_button = widgets.ButtonWidget(description="Quit")
quit_button.on_click(quit_mha)
display(quit_button)
```

Not that in this example, there is no complete GUI, because the redraw time would be too high when using the interactive widgets.  Therefor, the process is iterative:

1. Record the data from MHA.
2. Load the data from the recorded MAT file.
3. Plot the ColorMeshes.
4. Generate the interactive plot.
