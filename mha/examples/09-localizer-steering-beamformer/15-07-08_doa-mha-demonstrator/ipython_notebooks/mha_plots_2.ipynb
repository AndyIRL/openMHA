```python
from IPython.html import widgets # Widget definitions
from IPython.display import display # Used to display widgets in the notebook
import mpld3

mpld3.enable_notebook()

def plot_gcc_and_p(n1, n2):
    #subplots?
    fig, ax = subplots(1, 2, figsize=(10, 5))
    ax[0].plot(vGCC_mha[n1], '-x', label='GCC['+str(n1)+'] (MHA)')
    ax[0].plot(vGCC_ref[n2], '-o', label='GCC['+str(n2)+'] (Matlab)')
    ax[0].legend();

    ax[1].plot(p_mha[n1], '-x', label='p['+str(n1)+'] (MHA)')
    ax[1].plot(p_ref[n2], '-o', label='p['+str(n2)+'] (Matlab)')
    ax[1].legend()

widgets.interact(
    plot_gcc_and_p,
    n1=widgets.IntSliderWidget(min=1, max=vGCC_mha.shape[0]),
    n2=widgets.IntSliderWidget(min=1, max=vGCC_ref.shape[0])
);
```
